# Difference in numerical solutions

In this notebook we study how MATLAB's `ode45` compares to Python's `scipy.integrate.solve_ivp`. The code for data generation in both languages is available in `data_generation.py` and `data_generation.m`.

```python
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

py_data = loadmat('py_odes.mat')
oct_data = loadmat('oct_odes.mat')
```

```python
for ode in oct_data.keys():
    if ode not in ['t', 'lorenz']:
        difference = np.sum(py_data[ode] - oct_data[ode], axis=-1)
        plt.plot(difference, label=ode)
plt.legend()
plt.show()
```

```python
plt.plot(np.sum(py_data['lorenz'] - oct_data['lorenz'], axis=-1), label='lorenz')
plt.legend()
plt.show()
```

```python
data = oct_data['lorenz']
data2 = py_data['lorenz']

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(data[:,0], data[:,1], data[:,2], label='oct')
ax.plot(data2[:,0], data2[:,1], data2[:,2], label='py')
plt.legend()
plt.show()
```

```python
data = oct_data['mean_field']
data2 = py_data['mean_field']

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(data[:,0], data[:,1], data[:,2], label='oct')
ax.plot(data2[:,0], data2[:,1], data2[:,2], label='py')
plt.legend()
plt.show()
```

```python
py = py_data['van_der_pol']
oc = oct_data['van_der_pol']

plt.plot(py[:, 0], label='py_x'); plt.plot(py[:, 1], label='py_y')
plt.plot(oc[:, 0], label='oc_x'); plt.plot(oc[:, 1], label='oct_y')
plt.legend()
plt.show()
```

```python
py = py_data['linear_oscillator']
oc = oct_data['linear_oscillator']

plt.plot(py[:, 0], label='py_x'); plt.plot(py[:, 1], label='py_y')
plt.plot(oc[:, 0], label='oct_x'); plt.plot(oc[:, 1], label='oct_y')
plt.legend()
plt.show()
```

```python
py = py_data['mean_field']
oc = oct_data['mean_field']

plt.plot(py[:, 0], label='py_x'); plt.plot(py[:, 1], label='py_y')
plt.plot(oc[:, 0], label='oc_x'); plt.plot(oc[:, 1], label='oct_y')
plt.legend()
plt.show()
```
