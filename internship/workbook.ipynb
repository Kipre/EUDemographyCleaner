```python
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

py_data = loadmat('py_odes.mat')
oct_data = loadmat('oct_odes.mat')
```

```python
oct_data.keys()
```

```python
for ode in oct_data.keys():
    difference = np.sum(py_data[ode] - oct_data[ode], axis=-1)
    plt.plot(difference, label=ode)
plt.legend()
plt.show()
```

```python
data = oct_data['lorenz']
data2 = py_data['lorenz']

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(data[:,0], 
        data[:,1], 
        data[:,2],
        label='oct')
ax.plot(data2[:,0], 
        data2[:,1], 
        data2[:,2],
        label='oct')
plt.show()
```
